#  Analysis & Interpretation

## 1. Comparison of Anomaly Detection Models

In this phase, anomaly detection results from three models were compared:

- **Isolation Forest**
- **Local Outlier Factor (LOF)**
- **One-Class SVM**

The number of anomalies detected by each model is as follows:

- Isolation Forest: **78 anomalies**
- LOF: **78 anomalies**
- One-Class SVM: **77 anomalies**

Although the total number of anomalies detected by each model is very similar, the exact data points identified differ slightly. This is because each model uses a different approach to anomaly detection.

---

## 2. Model-Specific Behavior

Each model captures a different type of anomaly:

- **Isolation Forest**  
  Detects **global anomalies** by isolating extreme values in the dataset. It is effective in identifying large spikes and sudden drops in case counts.

- **Local Outlier Factor (LOF)**  
  Detects **local anomalies** by comparing the density of a point to its neighbors. It can identify smaller irregularities that may not appear extreme globally.

- **One-Class SVM**  
  Learns the boundary of normal data and flags points outside this boundary as anomalies. It is sensitive to complex and non-linear patterns.

Because of these differences, each model identifies slightly different anomaly points.

---

## 3. Overlapping Anomalies (Common Anomalies)

To improve reliability, Anomalies detected by all three models were identified.

- **Common anomalies detected by all models: 32**

These overlapping anomalies are the most significant because:

- All three models agree on them
- They are unlikely to be random noise
- They represent strong deviations in the dataset

---

## 4. Interpretation of Results

### 4.1 Model-wise Comparison

- While all models detect a similar number of anomalies, the overlap is smaller (32), showing that each model captures different aspects of irregular behavior.
- Isolation Forest focuses on extreme values across the dataset.
- LOF identifies local inconsistencies between neighboring data points.
- One-Class SVM captures deviations from the overall learned pattern.

---

### 4.2 Common vs Individual Anomalies

Anomalies can be categorized into:

- **Common anomalies (32 points)**  
  Detected by all three models → **most reliable**

- **Model-specific anomalies (remaining points)**  
  Detected by only one or two models → **less reliable but still informative**

---

### 4.3 Real-World Meaning

The detected anomalies correspond to:

- Sudden spikes in COVID-19 case counts
- Sharp drops following spikes
- Deviations from expected trends (rolling average)

These patterns may be caused by:

- Outbreak surges (e.g., COVID waves)
- Reporting delays
- Backlog updates or corrections

---

## 6. Key Insights

- All three models detected a similar number of anomalies, confirming consistency.
- Only **32 anomalies overlap**, indicating strong agreement across models.
- Different models capture different types of anomalies (global, local, boundary-based).
- Combining multiple models improves robustness and reliability.
- Common anomalies represent the most meaningful irregularities in the dataset.

---

## 7. Conclusion

The comparison of multiple anomaly detection models demonstrates that no single model captures all irregularities. Each model contributes a unique perspective, and their combination provides a more comprehensive understanding of the data.

The identification of overlapping anomalies strengthens confidence in the results and ensures that the most critical anomalies are accurately detected.

Overall, this multi-model approach provides a reliable framework for detecting unusual patterns in large-scale COVID-19 data.
