```python
import pandas as pd

#File paths:
input_file = "COVID-19_Case_Surveillance_Public_Use_Data_20260319.csv" #this will vary depending on what you named the original copy of the original csv file
output_file = "0.5_cdc_case_surveillance_sample.csv"

#Settings:
chunk_size = 100000
sample_frac = 0.005   # 0.5% sample from each chunk
random_seed = 42

sampled_chunks = []

for i, chunk in enumerate(pd.read_csv(input_file, chunksize=chunk_size)):
    # Clean column names
    chunk.columns = chunk.columns.str.strip().str.lower()

    # Making sure expected date column exists
    if 'cdc_case_earliest_dt' not in chunk.columns:
        print(f"Skipping chunk {i+1}: missing 'cdc_case_earliest_dt'")
        print("Columns found:", chunk.columns.tolist())
        continue

    # Converting date column to datetime
    chunk['cdc_case_earliest_dt'] = pd.to_datetime(
        chunk['cdc_case_earliest_dt'],
        errors='coerce'
    )

    # Drop rows without valid dates
    chunk = chunk.dropna(subset=['cdc_case_earliest_dt'])

    # Skipping empty chunks after dropping bad dates
    if len(chunk) == 0:
        continue

    # Taking samples from each chunk
    chunk_sample = chunk.sample(frac=sample_frac, random_state=random_seed)

    sampled_chunks.append(chunk_sample)

    print(f"Finished chunk {i+1} | original rows: {len(chunk)} | sampled rows: {len(chunk_sample)}")

# Combine all sampled chunks
sample_df = pd.concat(sampled_chunks, ignore_index=True)

# Save sample
sample_df.to_csv(output_file, index=False)

print("\nDone.")
print(f"Sample saved as: {output_file}")
print(f"Final sampled rows: {len(sample_df)}")
```
